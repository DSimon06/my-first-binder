### _Partie 1_ : Evolution des quantités de matière

Le programme ci-dessous permet de tracer l'évolution des quantités de matière lors du titrage de l'acide méthanoïque par une solution d'hydroxyde de sodium.

>Q1- Exécuter le programme et exploiter la courbe pour déterminer le volume versé à l'équivalence.

>Q2- Retrouver la ligne de programmation où est calculé le volume versé à l'équivalence et ajouter la note `# Calcul de Véquiv`.

```python
import numpy as np
from matplotlib import pyplot as plt

Ca= float(input("Entrer la valeur Ca de la concentration de l'acide à titrer en mol/L : Ca = "))
Va= float(input("Entrer le volume Va d'acide titré en mL : Va = "))
Cb= float(input("Entrer la valeur de la concentration Cb de la base contenue dans la burette graduée en mol/L : Cb = "))

N=40 # Nombre de points de la courbe
V=25 # Volume de la burette graduée en mL

Vb=[]
naf=[]
nbf=[]
Vbe=Ca*Va/Cb

for i in range (0,N):
    Vb.append(V*i/N)

    if Vb[i]< Vbe :
        nbf.append(0)
        naf.append(Ca*Va/1000-Cb*Vb[i]/1000)

    else :
        nbf.append(Cb*Vb[i]/1000-Ca*Va/1000)
        naf.append(0)

plt.title("Evolution des quantités d'acide méthanoïque et d'ions hydroxyde au cours du titrage")
plt.xlabel("Volume de solution titrante versé Vb(mL)")
plt.ylabel("Quantités de matière en mol")

plt.grid(True)

plt.plot(Vb,naf,c='red',label="Quantité d'acide méthanoïque présent dans le bécher : na(mol)")
plt.plot(Vb,nbf,c='green',label="Quantité d'ions hydroxyde présents dans le bécher : nb(mol")
plt.legend()
plt.show()
```

```python
Ca= float(input("Entrer la valeur Ca de la concentration de l'acide à titrer en mol/L : Ca = "))
Va= float(input("Entrer le volume Va d'acide titré en mL : Va = "))
Cb= float(input("Entrer la valeur de la concentration Cb de la base contenue dans la burette graduée en mol/L : Cb = "))

N=40 # Nombre de points de la courbe
V=25 # Volume de la burette graduée en mL

Vb=[]
naf=[]
nbf=[]
ncf=[]

Vbe=3*Ca*Va/Cb

for i in range (0,N):
    Vb.append(V*i/N)

    if Vb[i]< Vbe :
        nbf.append(0)
        naf.append(Ca*Va/1000-Cb*Vb[i]/3/1000)
        ncf.append(Cb*Vb[i]/3/1000)       
        
    else :
        nbf.append(Cb*Vb[i]/1000-3*Ca*Va/1000)
        naf.append(0)
        ncf.append(Cb*Vbe/3/1000)
       
plt.title("Evolution des quantités de matière au cours du titrage")
plt.xlabel("Volume de solution titrante versé Vb(mL)")
plt.ylabel("Quantités de matière en mol")

plt.grid(True)

plt.plot(Vb,naf,c='red',label="Quantité d'acide citrique présent dans le bécher : na(mol)")
plt.plot(Vb,nbf,c='green',label="Quantité d'ions hydroxyde présents dans le bécher : nb(mol")
plt.plot(Vb,ncf,c='blue',label="Quantité d'ions citrate présents dans le bécher : nc(mol")

plt.legend()
plt.show()
```

### _Partie 2_ : Evolution de la conductivité

On souhaite représenter l'évolution de la conductivité du mélange contenu dans le bécher au fur et à mesure du titrage.

Dans le programme suivant,  on a ajouté les lignes de programmation nécessaires pour créer et calculer la quantité de matière nd d'ions sodium présents dans le bécher au cours du titrage, ainsi que la conductivité sigma du mélange.

>Q9- Pourquoi le calcul de nd, dans la boucle `for`, est-il placé en dehors de la boucle conditionnelle `if`, liée au volume versé à l'équivalence ?

>Q10- Pourquoi, dans le cadre de l'étude de la conductivité, a-t-on supprimé les lignes de programmation concernant l'acide citrique (na) ?

>Q11- Exécuter le programme. Comment repère-t-on l'équivalence ?

```python
Ca= float(input("Entrer la valeur Ca de la concentration de l'acide à titrer en mol/L : Ca = "))
Va= float(input("Entrer le volume Va d'acide titré en mL : Va = "))
Cb= float(input("Entrer la valeur de la concentration Cb de la base contenue dans la burette graduée en mol/L : Cb = "))

N=40 # Nombre de points de la courbe
V=25 # Volume de la burette graduée en mL

# Conductivités molaires ioniques en S.m²/mol :
Lambdab=0.0199
Lambdac=0.0210
Lambdad=0.0050

Vb=[]
nbf=[]
ncf=[]
ndf=[]
sigma=[]

Vbe=3*Ca*Va/Cb

for i in range (0,N):
    Vb.append(V*i/N)
    ndf.append(Cb*Vb[i]/1000)

    if Vb[i]< Vbe :
        nbf.append(0)        
        ncf.append(Cb*Vb[i]/3/1000)
        sigma.append(1/((Va+Vb[i])*1000*1000)*(Lambdab*nbf[i]+Lambdac*ncf[i]+Lambdad*ndf[i]))
        
    else :
        nbf.append(Cb*Vb[i]/1000-3*Ca*Va/1000)        
        ncf.append(Cb*Vbe/3/1000)
        sigma.append(1/((Va+Vb[i])*1000*1000)*(Lambdab*nbf[i]+Lambdac*ncf[i]+Lambdad*ndf[i]))

plt.title("Evolution de la conductivité au cours du titrage")
plt.xlabel("Volume de solution titrante versé Vb(mL)")
plt.ylabel("Conductivité en S/m")

plt.grid(True)

plt.plot(Vb,sigma,c='red')

plt.show()
```

Le calcul de la conductivité sigma commence par le facteur `1/((Va+Vb[i])*1000*1000)`.
