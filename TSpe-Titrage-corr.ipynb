### _Partie 1_ : Evolution des quantités de matière

Le programme ci-dessous permet de tracer l'évolution des quantités de matière lors du titrage de l'acide méthanoïque par une solution d'hydroxyde de sodium.

>Q1- Exécuter le programme et exploiter la courbe pour déterminer le volume versé à l'équivalence.  
***Véquiv = 4 mL***

>Q2- Retrouver la ligne de programmation où est calculé le volume versé à l'équivalence et ajouter la note `# Calcul de Véquiv`.

```python
import numpy as np
from matplotlib import pyplot as plt

Ca= float(input("Entrer la valeur Ca de la concentration de l'acide à titrer en mol/L : Ca = "))
Va= float(input("Entrer le volume Va d'acide titré en mL : Va = "))
Cb= float(input("Entrer la valeur de la concentration Cb de la base contenue dans la burette graduée en mol/L : Cb = "))

N=40 # Nombre de points de la courbe
V=25 # Volume de la burette graduée en mL

Vb=[]
naf=[]
nbf=[]
Vbe=Ca*Va/Cb # Calcul de Véquiv

for i in range (0,N):
    Vb.append(V*i/N)

    if Vb[i]< Vbe :
        nbf.append(0) # nb avant l'équivalence
        naf.append(Ca*Va/1000-Cb*Vb[i]/1000) # na avant l'équivalence

    else :
        nbf.append(Cb*Vb[i]/1000-Ca*Va/1000) # nb après l'équivalence
        naf.append(0) # na après l'équivalence

plt.title("Evolution des quantités d'acide méthanoïque et d'ions hydroxyde au cours du titrage")
plt.xlabel("Volume de solution titrante versé Vb(mL)")
plt.ylabel("Quantités de matière en mol")

plt.grid(True)

plt.plot(Vb,naf,c='red',label="Quantité d'acide méthanoïque présent dans le bécher : na(mol)")
plt.plot(Vb,nbf,c='green',label="Quantité d'ions hydroxyde présents dans le bécher : nb(mol")
plt.legend()
plt.show()
```

>Q3- La réaction étant très rapide, les réactifs réagissent instantanément et l'état final est atteint pour chaque valeur de volume versé. Justifier alors l'allure de la courbe de nb avant l'équivalence. Retrouver la ligne de programmation correspondante et ajouter la note `# nb avant l'équivalence`.  
***Avant l'équivalence, tous les ions hydroxyde sont consommés dès qu'on les ajoute car ils sont limitants. Leur quantité de matière est donc toujours nulle à l'état final, tant qu'on est avant l'équivalence***

>Q4- De même, justifier l'allure de la courbe et annoter les lignes concernant nb après l'équivalence, et na avant et après l'équivalence.  
***nb après l'équivalence : les ions hydroxyde sont ajoutés, mais il n'y a plus d'acide citrique. Donc aucune réaction n'a plus lieu et la quantité d'ions hydroxyde augmente.  
na avnt l'équivalence : la quantité d'acide citrique diminue au fur et à mesure de l'ajout de la solution titrante, car ils réagissent avec les ions hydroxyde ajoutés.
na après l'équivalence : à l'équivalence, tout l'acide citrique est consommé. Donc à parir de ce moment, il n'y en a plus.***

>Q6- Adapter le calcul du volume à l'équivalence, établi grâce au tableau d'avancement.  
***CaVa = CbVb/3 donc à l'équivalence Vbe = 3CaVa/Cb***

>Q7- Adapter si besoin les calculs de na et nb avant et après l'équivalence, grâce aux questions de l'activité et exécuter le programme pour vérifier la cohérence des courbes obtenues.  
***Avant l'équivalence : na = CaVa-x et comme les ions hydroxyde sont limitants : CbVb-3x = 0 donc x = CbVb/3 ; d'où na = CaVa-CbVb/3  
Après l'équivalence : nb = tout ce qu'on a ajouté (CbVb) - ce qui a réagi avec tout l'acide citrique (3CaVa)***

>Q8- Ajouter les lignes de programmation nécessaires pour créer, calculer et tracer la quantité de matière nc d'ions citrate présents dans le bécher au cours du titrage.  
***On doit créer la grandeur `ncf=[]`,  
puis la calculer : avant l'équivalence nc = x = CbVb/3 ; et après l'équivalence tout l'acide citrique a été transformé en ions citrate et la quantité d'ions citrate reste constante, égale à CaVa.  
Puis il faut tracer la courbe avec la commande `plt.plot`complétée comme celles de na et nb.***


```python
Ca= float(input("Entrer la valeur Ca de la concentration de l'acide à titrer en mol/L : Ca = "))
Va= float(input("Entrer le volume Va d'acide titré en mL : Va = "))
Cb= float(input("Entrer la valeur de la concentration Cb de la base contenue dans la burette graduée en mol/L : Cb = "))

N=40 # Nombre de points de la courbe
V=25 # Volume de la burette graduée en mL

Vb=[]
naf=[]
nbf=[]
ncf=[]

Vbe=3*Ca*Va/Cb # Calcul du volume à l'équivalence

for i in range (0,N):
    Vb.append(V*i/N)

    if Vb[i]< Vbe :
        nbf.append(0)
        naf.append(Ca*Va/1000-Cb*Vb[i]/3/1000)
        ncf.append(Cb*Vb[i]/3/1000)
        
    else :
        nbf.append(Cb*Vb[i]/1000-3*Ca*Va/1000)
        naf.append(0)
        ncf.append(Ca*Va/1000)
       
plt.title("Evolution des quantités de matière au cours du titrage")
plt.xlabel("Volume de solution titrante versé Vb(mL)")
plt.ylabel("Quantités de matière en mol")

plt.grid(True)

plt.plot(Vb,naf,c='red',label="Quantité d'acide citrique présent dans le bécher : na(mol)")
plt.plot(Vb,nbf,c='green',label="Quantité d'ions hydroxyde présents dans le bécher : nb(mol)")
plt.plot(Vb,ncf,c='blue',label="Quantité d'ions citrate présents dans le bécher : nc(mol)")

plt.legend()
plt.show()
```

### _Partie 2_ : Evolution de la conductivité

On souhaite représenter l'évolution de la conductivité du mélange contenu dans le bécher au fur et à mesure du titrage.

Dans le programme suivant,  on a ajouté les lignes de programmation nécessaires pour créer et calculer la quantité de matière nd d'ions sodium présents dans le bécher au cours du titrage, ainsi que la conductivité sigma du mélange.

>Q9- Pourquoi le calcul de nd, dans la boucle `for`, est-il placé en dehors de la boucle conditionnelle `if`, liée au volume versé à l'équivalence ?  
***Les ions sodium ne participent pas à la réaction, ils sont spectateurs. Donc leur quantité augmente de la même façon avant et après l'équivalence. Elle ne dépend donc pas de la condition sur Vb.***

>Q10- Pourquoi, dans le cadre de l'étude de la conductivité, a-t-on supprimé les lignes de programmation concernant l'acide citrique (na) ?  
***Seuls les ions participent à la conductivité. L'acide citrique n'est pas chargé et n'intervient donc pas dans le calcul de la conductivité.***

>Q11- Exécuter le programme. Comment repère-t-on l'équivalence ?  
***L'équivalence correspond à l'intersection entre les deux portions de courbe.***

```python
Ca= float(input("Entrer la valeur Ca de la concentration de l'acide à titrer en mol/L : Ca = "))
Va= float(input("Entrer le volume Va d'acide titré en mL : Va = "))
Cb= float(input("Entrer la valeur de la concentration Cb de la base contenue dans la burette graduée en mol/L : Cb = "))

N=40 # Nombre de points de la courbe
V=25 # Volume de la burette graduée en mL

# Conductivités molaires ioniques en S.m²/mol :
Lambdab=0.0199
Lambdac=0.0210
Lambdad=0.0050

Vb=[]
nbf=[]
ncf=[]
ndf=[]
sigma=[]

Vbe=3*Ca*Va/Cb

for i in range (0,N):
    Vb.append(V*i/N)
    ndf.append(Cb*Vb[i]/1000)

    if Vb[i]< Vbe :
        nbf.append(0)        
        ncf.append(Cb*Vb[i]/3/1000)
        sigma.append(1/((Va+Vb[i]+200)*1000*1000)*(Lambdab*nbf[i]+Lambdac*ncf[i]+Lambdad*ndf[i]))
        
    else :
        nbf.append(Cb*Vb[i]/1000-3*Ca*Va/1000)        
        ncf.append(Cb*Vbe/3/1000)
        sigma.append(1/((Va+Vb[i]+200)*1000*1000)*(Lambdab*nbf[i]+Lambdac*ncf[i]+Lambdad*ndf[i]))

plt.title("Evolution de la conductivité au cours du titrage")
plt.xlabel("Volume de solution titrante versé Vb(mL)")
plt.ylabel("Conductivité en S/m")

plt.grid(True)

plt.plot(Vb,sigma,c='red')

plt.show()
```

Le calcul de la conductivité sigma commence par le facteur `1/((Va+Vb[i])*1000*1000)`.

>Q12- Que représente `Va+Vb[i]` et pourquoi est-il multiplié deux fois par 1000 ?  
***C'est le volume total de la solution, conveti en m<sup>3</sup>.***
